```python
def AbsoluteSubtract(Seq1, Seq2):
    '''Seq1 and Seq2 is a numpy array, should be a (3,M,N)picture
    Output = int'''
    assert Seq1.shape == Seq2.shape, "Improper shapes"
    Seq1 = Seq1.flatten()
    Seq2 = Seq2.flatten()
    Seq = np.abs(Seq1-Seq2)
    Seq = Seq/(255*255*255)
    print(f"type of Sequnce is type{Seq} before all")
    
    Seq = np.sum(Seq)
    Seq = 1-Seq
    
    
    print(f"seq1 is {Seq1}")
    print(f"Seq2 is {Seq2}")
    #tot=3*Seq1
    print(f"type of Sequnce is type{Seq}")
    print (f"Sequence is {Seq}")
    return Seq
AbsSum = np.vectorize(AbsoluteSubtract)
```

```python
def DeleteFormer(Captcha, Filter, zero = 255):
    '''This removes the cross ection of the Filter and Captcha so that they arent repeated
    Input 
    Captcha.size = Filter.size
    Zero is the value of white which we will ignore
    
    Output
    Captcha without Filter'''
    
    assert isinstance(Captcha, np.ndarray) and isinstance(Filter, np.ndarray), f"The inputs must me np.ndarrays, the Captcha type is {type(Captcha)} and the Filter type is {type(Filter)} "
    assert Captcha.shape == Filter.shape, f"Your Filter and Captcha must be the same size, this is the filter {Filter.shape} and the Captcha is {Captcha.shape}"
    
    for i in np.arange(Captcha.shape[0]):
        for j in np.arange(Captcha.shape[1]):
            for k in np.arange(Captcha.shape[2]):
                if Filter[i,j,k]!= zero and Captcha[i,j,k]!=zero:
                    Captcha[i,j,k]=zero
    return zero
```

```python
def FilterCaptcha(Captcha, FiltDic, Hyperparameter, Stride =(1,1)):
    '''Takes in captcha and finds the best match of each letter by comparing similarities and finding the most similar
    Input: Captcha, Filter, Hyperparameter
    Captcha is an np.ndarray of shape (H,W,C)   M is y N is x
    Filter is a list of (np.ndarray of shape (H,W,C) converted toz(3,FM, FN))
    FiltDic is a list with the filter picture array as the first element in the tuple and representation as second
    Hyperparameter is the minimum similarity or sequence
    Stride is the movement
    
    OutPut: strings(A string of the Captcha)'''
    #Captcha = SimplifyCaptcha(Captcha)
    
    Captcha=Captcha.transpose(2,1,0)
    print(f"Captcha.size is {Captcha.shape}")
    
    for i in np.arange(len(FiltDic)):
        #FiltDic[i][0]=SimplifyCaptcha(FiltDic[i][0]).transpose(2,0,1)
        FiltDic[i]=(FiltDic[i][0].transpose(2,1,0),FiltDic[i][1])
    assert isinstance(Captcha, np.ndarray) and len(Captcha.shape)==3 and 3 in Captcha.shape, "Your Captcha is wrong"
    assert Captcha.shape[1] >= FiltDic[0][0].shape[1], "Your filter width is larger than your file, this is a big problem" 
    assert Captcha.shape[2] >= FiltDic[0][0].shape[2], f"Your filter height is larger than your file, this is a big problem\nYour fh is {FiltDic[0][0].shape[2]} and ch is{Captcha.shape}"
    assert isinstance(FiltDic, list), "FiltDic is not a dictionary"
    
    word =""
    #HorizontalMax = (Captcha.shape[2]//FiltDic.keys()[0].shape[2])*Filter.shape[2]
    #VerticalMax = (Captcha.shape[1]//FiltDic.keys()[0])*Filter.shape[1]
    #Filters = np.array(Filter)
    
    i =1
    j =1
    while(i<=Captcha.shape[1]-FiltDic[0][0].shape[1]):
        while(j<=Captcha.shape[2]-FiltDic[0][0].shape[2]):
            mostlikely = ''
            for Filterz in FiltDic:
                print(f"Filter is {Filterz[0][:,:,:].shape} Captcha is {Captcha[:,i:i+Filterz[0].shape[1],j:j+Filterz[0].shape[2]].shape}")
                if (AbsoluteSubtract(Filterz[0],Captcha[:,i:i+Filterz[0].shape[1],j:j+Filterz[0].shape[2]]) >= Hyperparameter):
                    print(f"Match with {Filterz[-1]} with {AbsoluteSubtract(Filterz[0],Captcha[:,i:i+Filterz[0].shape[1],j:j+Filterz[0].shape[2]])}")
                    Captcha[:,i:i+Filterz[0].shape[1],j:j+Filterz[0].shape[2]] = DeleteFormer(Captcha[:,i:i+Filterz[0].shape[1],j:j+Filterz[0].shape[2]],  Filterz[0])
                    mostlikely=Filterz[-1]
                else:
                    print(f"Filterz[-1], is a {AbsoluteSubtract(Filterz[0],Captcha[:,i:i+Filterz[0].shape[1],j:j+Filterz[0].shape[2]])}")
            if mostlikely!='':
                word+=mostlikely
            j+=Stride[-1]
        i+=Stride[0]
        
    
    
    return word
    
    
```
