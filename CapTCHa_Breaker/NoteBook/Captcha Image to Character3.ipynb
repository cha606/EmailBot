```python
def AbsoluteSubtract(Seq1, Seq2):
    '''Seq1 and Seq2 is a numpy array, should be a (3,M,N)picture
    Output = int'''
    assert Seq1.shape == Seq2.shape, "Improper shapes"
    Seq = np.abs(Seq1-Seq2)
    print(Seq1)
    print(Seq2)
    tot=3*Seq1
    return np.sum(Seq)/tot
AbsSum = np.vectorize(AbsoluteSubtract)
```

```python
def FilterCaptcha(Captcha, FiltDic, Hyperparameter, Stride =(1,1)):
    '''Takes in captcha and finds the best match of each letter by comparing similarities and finding the most similar
    Input: Captcha, Filter, Hyperparameter
    Captcha is an np.ndarray of shape (H,W,C)   M is y N is x
    Filter is a list of (np.ndarray of shape (H,W,C) converted toz(3,FM, FN))
    FiltDic is a list with the filter picture array as the first element in the tuple and representation as second
    Hyperparameter is the minimum similarity or sequence
    Stride is the movement
    
    OutPut: strings(A string of the Captcha)'''
    #Captcha = SimplifyCaptcha(Captcha)
    
    Captcha=Captcha.transpose(2,0,1)
    print(f"Captcha.size is {Captcha.shape}")
    
    for i in np.arange(len(FiltDic)):
        #FiltDic[i][0]=SimplifyCaptcha(FiltDic[i][0]).transpose(2,0,1)
        FiltDic[i]=(FiltDic[i][0].transpose(0,1,2),FiltDic[i][1])
    assert isinstance(Captcha, np.ndarray) and len(Captcha.shape)==3 and 3 in Captcha.shape, "Your Captcha is wrong"
    assert Captcha.shape[1] >= FiltDic[0][0].shape[1], "Your filter width is larger than your file, this is a big problem" 
    assert Captcha.shape[2] >= FiltDic[0][0].shape[2], f"Your filter height is larger than your file, this is a big problem\nYour fh is {FiltDic[0][0].shape[2]} and ch is{Captcha.shape}"
    assert isinstance(FiltDic, list), "FiltDic is not a dictionary"
    
    word =""
    #HorizontalMax = (Captcha.shape[2]//FiltDic.keys()[0].shape[2])*Filter.shape[2]
    #VerticalMax = (Captcha.shape[1]//FiltDic.keys()[0])*Filter.shape[1]
    #Filters = np.array(Filter)
    
    i =1
    j =1
    while(i<=Captcha.shape[1]-FiltDic[0][0].shape[1]):
        while(j<=Captcha.shape[2]-FiltDic[0][0].shape[2]):
            mostlikely = ''
            for Filterz in FiltDic:
                print(f"Filter is {Filterz[0][:,:,:].shape} Captcha is {Captcha[i:i+Filterz[0].shape[2],j:j+Filterz[0].shape[1],:].shape}")
                if (AbsSum(Filterz[0][:,:,:],Captcha[i:i+Filterz[0].shape[2],j:j+Filterz[0].shape[1],:]) >= Hyperparameter):
                    mostlikely=Filterz[-1]
            if mostlikely!='':
                word+=mostlikely
            j+=Stride[-1]
        i+=Stride[0]
        
    
    
    return word
    
    
```
